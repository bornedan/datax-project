# Preprocesing
For our project we are used only listing.csv dataset. This dataset contains text, numerical, categorical, boolean, date, id and null columns.
39 columns (52 %) contains some null values and 6 columns (8 %) contains only null values.
## Column type:
Text Columns: listing_url, name, neighborhood_overview, picture_url, host_url, host_name, host_location, host_about, host_thumbnail_url, host_picture_url, host_neighbourhood, neighbourhood, property_type, bathrooms_text

Categorical Columns: source, host_response_time, host_verifications, neighbourhood_cleansed, room_type

Null colums: description, neighbourhood_group_cleansed, bathrooms, bedrooms,calendar_updated, license, amenites

Datetime columns: last_scraped, host_since, calendar_last_scraped, first_review, last_review

Id columns: id, host_id, scrape_id

Boolean columns: host_is_superhost, host_has_profile_pic, host_identity_verified, has_availability, instant_bookable

```python
null_percentage = (df.isnull().mean() * 100).round(2)

# Filter null_percentage to include only columns with null values greater than zero
null_percentage_greater_than_zero = null_percentage[null_percentage > 0]

# Create a bar plot of null values percentage
plt.figure(figsize=(10, 6))
sns.barplot(x=null_percentage_greater_than_zero.values, y=null_percentage_greater_than_zero.index, palette='viridis')
plt.xlabel('Percentage of Null Values')
plt.ylabel('Column names')
plt.title('Percentage of Null Values when is greater then zero')
plt.show()
```
